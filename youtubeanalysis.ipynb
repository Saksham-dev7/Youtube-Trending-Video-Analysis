```python
import pandas as pd

df = pd.read_csv('USvideos.csv')
display(df.head())
```

```python
import pandas as pd
import re


for col in ['views', 'likes', 'dislikes', 'comment_count']:
    if df[col].isnull().any():
        df[col] = df[col].fillna(0).astype(int)

for col in ['tags', 'title', 'description']:
    if df[col].isnull().any():
        df[col] = df[col].fillna('')


df.drop_duplicates(inplace=True)


df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m', errors='coerce')
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')


def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation and special characters
    return text

for col in ['tags', 'title', 'description']:
    df[col] = df[col].apply(clean_text)

display(df.head())
```

```python
# Calculate descriptive statistics for 'views', 'likes', and 'comment_count'
numerical_stats = df[['views', 'likes', 'comment_count']].describe()
display(numerical_stats)

# Analyze the distribution of videos across different categories
if 'category_name' in df.columns:
    category_counts = df['category_name'].value_counts()
else:
    category_counts = df['category_id'].value_counts()
    print("Note: Using 'category_id' instead of 'category_name' because the latter is not available.")
display(category_counts)

# Calculate the correlation matrix for 'views', 'likes', and 'comment_count'
correlation_matrix = df[['views', 'likes', 'comment_count']].corr()
display(correlation_matrix)
```

```python
# Group data by category and calculate aggregate statistics.
if 'category_name' in df.columns:
    category_stats = df.groupby('category_name').agg(
        {'video_id': 'count', 'views': 'mean', 'likes': 'mean'}
    )
else:
    category_stats = df.groupby('category_id').agg(
        {'video_id': 'count', 'views': 'mean', 'likes': 'mean'}
    )
    print("Note: Using 'category_id' instead of 'category_name' because the latter is not available.")
display(category_stats)
```

```python
import matplotlib.pyplot as plt

# 1. Average views per category
plt.figure(figsize=(12, 6))
category_stats['views'].plot(kind='bar', color='skyblue')
plt.title('Average Views per Category (using category_id)')
plt.xlabel('Category ID')
plt.ylabel('Average Views')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Count of trending videos per category
plt.figure(figsize=(12, 6))
category_stats['video_id'].plot(kind='bar', color='lightcoral')
plt.title('Count of Trending Videos per Category (using category_id)')
plt.xlabel('Category ID')
plt.ylabel('Number of Trending Videos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Boxplot of views distribution per category
plt.figure(figsize=(12, 6))
df.boxplot(column='views', by='category_id', grid=False, figsize=(12, 8), patch_artist=True, showfliers=False)  # Hide outliers to improve visualization
plt.title('Distribution of Views per Category (using category_id)')
plt.xlabel('Category ID')
plt.ylabel('Views')
plt.suptitle('') # Remove default boxplot title
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Define a list of clickbait words
clickbait_words = ["shocking", "amazing", "unbelievable", "incredible", "mind-blowing", "astounding", "epic", "insane", "must watch", "you won't believe", "secret", "revealed", "top 10", "best ever", "life changing"]

# Create a new column 'clickbait_score' and initialize it to 0
df['clickbait_score'] = 0


for index, row in df.iterrows():
    title = row['title']
    for word in clickbait_words:
        if word in title:
            df.loc[index, 'clickbait_score'] += 1

display(df.head())
```

```python
# Calculate the median clickbait_score
median_clickbait_score = df['clickbait_score'].median()

# Group data by high/low clickbait score
grouped_data = df.groupby(df['clickbait_score'] > median_clickbait_score)['views'].mean()
high_clickbait_avg_views = grouped_data[True]
low_clickbait_avg_views = grouped_data[False]


print(f"Average views for high clickbait score videos: {high_clickbait_avg_views}")
print(f"Average views for low clickbait score videos: {low_clickbait_avg_views}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(['High Clickbait', 'Low Clickbait'], [high_clickbait_avg_views, low_clickbait_avg_views], color=['skyblue', 'lightcoral'])
plt.xlabel('Clickbait Score')
plt.ylabel('Average Views')
plt.title('Average Views for High vs. Low Clickbait Score Videos')
plt.show()
```
